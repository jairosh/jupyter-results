```python
import pandas as pd

pd.set_option('display.max_colwidth', -1)
df = pd.read_csv('maxprop_files.csv', parse_dates=['accessed'])
df2 = pd.read_csv('maxprop_files_2.csv', parse_dates=['accessed'])
df = df.append(df2)
display(df)
```

```python
allfiles = df.shape[0]
df = df[df.fsize > 0]
nonempty = df.shape[0]
print("All files: {0}, non empty files: {1}".format(allfiles, nonempty))
df = df[df.buffer_size == 50]
```

```python
for group,s in df.groupby(by=['nodes', 'message_interval']):
    print("{}-{}".format(group, s.shape[0]))
    
```

```python
new = pd.DataFrame()
for group, s in df.groupby(by=['file']):
    if s.shape[0] > 1:
        #print("{} has {}".format(group, s.shape[0]))
        #display(s)
        newer = s[s['accessed'] == s['accessed'].min()]
        new = new.append(newer)
    else:
        new = new.append(s)    
display(new)
new.describe()
```

```python
all_seeds = [78802, 94176, 65221, 44180, 79009, 68831, 73906, 15598, 78051, 71182, 15475, 45907, 5768, 28749, 24113, 80454, 59004, 54008, 70, 8439, 22, 58405, 38517, 33282, 5481, 11778, 96617, 52770, 55182, 5327, 11131, 61959, 44450, 4762, 20098, 61667, 15405, 98682, 82509, 65139, 4029, 90521, 98820, 3136, 35402, 64524, 37578, 16966, 84181, 91016]
seeds = all_seeds[:30]
seeds_df = pd.DataFrame({'seed': seeds})
#display(seeds_df)
valid_seeds = pd.merge(new, seeds_df, how='inner', on='seed')
display(valid_seeds)
```

```python
final_group = pd.DataFrame()
for group, s in valid_seeds.groupby(by=['nodes', 'message_interval']):
    print('{} has {}'.format(group, s.shape[0]))
    final_group = final_group.append(s)
```

```python
import os
def make_filepath(row):
    if any(row):
        full_path = os.path.join(row.folder, row.file)
        if row.host == 's30':
            return 'scp jairo@s30.local:\'{path}\' /home/jairo/Maxprop'.format(path=full_path)
        elif row.host == 'asterion':
            return 'cp \'{path}\' /home/jairo/Maxprop/{file}'.format(path=full_path, file=row.file)
        
    else:
        print("Ignored: {}".format(row))
        pass

x = final_group.apply(make_filepath, axis=1)
x.to_csv('cmds.csv', index=False)
```
