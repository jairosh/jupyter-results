```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
from matplotlib.backends.backend_pdf import PdfPages

ttl = pd.read_csv('/home/jairo/Simulaciones/CRAWDAD/ttl.csv')
means = ttl.groupby(by=['ttl', 'router']).mean()
means = means.unstack()
error = ttl.groupby(by=['ttl', 'router']).sem() * 1.96
error = error.unstack()
display(means)
```

```python
with PdfPages('/home/jairo/TTL_Response_infocomm.pdf') as pdf:
    ax = means['delivery_prob'].plot(figsize=(11,9.5))
    ax.set_xlabel('TTL (minutes)')
    ax.set_ylabel('Delivery Probability')
    handles, labels = ax.get_legend_handles_labels()
    lgd = plt.legend(handles, labels, loc='center left', bbox_to_anchor=(1.0, 0.5))
    pdf.savefig(bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.close()

    ax = means['overhead_ratio'].plot(figsize=(11,9.5))
    ax.set_xlabel('TTL (minutes)')
    ax.set_ylabel('Overhead ratio (copies created/messages delivered)')
    handles, labels = ax.get_legend_handles_labels()
    lgd = plt.legend(handles, labels, loc='center left', bbox_to_anchor=(1.0, 0.5))    
    pdf.savefig(bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.close()

    ax = means['latency_avg'].plot(figsize=(11,9.5))
    ax.set_xlabel('TTL (minutes)')
    ax.set_ylabel('Latency (s)')
    handles, labels = ax.get_legend_handles_labels()
    lgd = plt.legend(handles, labels, loc='center left', bbox_to_anchor=(1.0, 0.5))    
    pdf.savefig(bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.close()
```
